```python
dataset=dataset.drop('Route',axis=1)
```

```python
dataset['Total_Stops'].fillna('No Info',inplace=True)
```

```python
import datetime
import datetime
time_now=datetime.datetime.now()
now=time_now.time()
midn=now.replace(hour=0,minute=0,second=0,microsecond=0)
a6=now.replace(hour=6,minute=0,second=0,microsecond=0)
t12=now.replace(hour=12,minute=0,second=0,microsecond=0)
p6=now.replace(hour=18,minute=0,second=0,microsecond=0)
for i in range(len(dataset['Dep_Time'])):
    t2=time_now.time()
    h,m=dataset['Dep_Time'][i].split(':')
    t2=t2.replace(hour=int(h),minute=int(m),second=0,microsecond=0)
    if midn <= t2 < a6:
        dataset['Dep_Time'][i]='Class 1'
    elif a6 <= t2 < t12:
        dataset['Dep_Time'][i]='Class 2'
    elif t12 <= t2 < p6:
        dataset['Dep_Time'][i]='Class 3'
    else:
        dataset['Dep_Time'][i]='Class 4'
```

```python
for i in range(len(dataset['Arrival_Time'])):
    t3=time_now.time()
    time,*args=dataset['Arrival_Time'][i].split(' ')
    h,m=time.split(':')
    t3=t3.replace(hour=int(h),minute=int(m),second=0,microsecond=0)
    if midn <= t3 < a6:
        dataset['Arrival_Time'][i]='Class 1'
    elif a6 <= t3 < t12:
        dataset['Arrival_Time'][i]='Class 2'
    elif t12 <= t3 < p6:
        dataset['Arrival_Time'][i]='Class 3'
    else:
        dataset['Arrival_Time'][i]='Class 4'
```

```python
dataset['Duration']=dataset['Duration'].str.replace("h", '*60').str.replace(' ','+').str.replace('m','*1').apply(eval)
```

```python
dataset['Additional_Info']=dataset['Additional_Info'].map({'No info':1, 'In-flight meal not included':2,
       'No check-in baggage included':3, '1 Short layover':4, 'No Info':5,
       '1 Long layover':6, 'Change airports':7, 'Business class':8,
       'Red-eye flight':9, '2 Long layover':10})
```

```python
dataset['Airline']=dataset['Airline'].map({'IndiGo':1, 'Air India':2, 'Jet Airways':3, 'SpiceJet':4,
       'Multiple carriers':5, 'GoAir':6, 'Vistara':7, 'Air Asia':8,
       'Vistara Premium economy':9, 'Jet Airways Business':10,
       'Multiple carriers Premium economy':11, 'Trujet':12})
```

```python
import fastai
import re
def add_datepart(df, fldname, drop=True, time=True):
    
    fld = df[fldname]
    fld_dtype = fld.dtype
    if isinstance(fld_dtype, pd.core.dtypes.dtypes.DatetimeTZDtype):
        fld_dtype = np.datetime64

    if not np.issubdtype(fld_dtype, np.datetime64):
        df[fldname] = fld = pd.to_datetime(fld, infer_datetime_format=True)
    targ_pre = re.sub('[Dd]ate$', '', fldname)
    attr = ['Year', 'Month', 'Week', 'Day', 'Dayofweek', 'Dayofyear',
            'Is_month_end', 'Is_month_start', 'Is_quarter_end', 'Is_quarter_start', 'Is_year_end', 'Is_year_start']
    if time: attr = attr + ['Hour', 'Minute', 'Second']
    for n in attr: df[targ_pre + n] = getattr(fld.dt, n.lower())
    df[targ_pre + 'Elapsed'] = fld.astype(np.int64) // 10 ** 9
    if drop: df.drop(fldname, axis=1, inplace=True)
```

```python
add_datepart(dataset,'Date_of_Journey')
```

```python
dataset=dataset.drop('Date_of_JourneyYear',axis=1)
dataset=dataset.drop('Date_of_JourneyWeek',axis=1)
dataset=dataset.drop('Date_of_JourneyDayofyear',axis=1)
dataset=dataset.drop('Date_of_JourneyIs_quarter_end',axis=1)
dataset=dataset.drop('Date_of_JourneyIs_quarter_start',axis=1)
dataset=dataset.drop('Date_of_JourneyHour',axis=1)
dataset=dataset.drop('Date_of_JourneyMinute',axis=1)
dataset=dataset.drop('Date_of_JourneySecond',axis=1)
dataset=dataset.drop('Date_of_JourneyElapsed',axis=1)
dataset=dataset.drop('Date_of_JourneyIs_month_end',axis=1)
dataset=dataset.drop('Date_of_JourneyIs_month_start',axis=1)
dataset=dataset.drop('Date_of_JourneyIs_year_end',axis=1)
dataset=dataset.drop('Date_of_JourneyIs_year_start',axis=1)
```

```python
dataset['Source']=dataset['Source'].map({'Banglore':10, 'Kolkata':20, 'Delhi':30, 'Chennai':40, 'Mumbai':50})
dataset['Destination']=dataset['Destination'].map({'New Delhi':30, 'Banglore':10, 'Cochin':60, 'Kolkata':20, 'Delhi':30, 'Hyderabad':70})
```

```python
dataset['Total_Stops']=dataset['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4 , 'No Info': -999})
```

```python
dataset['Dep_Time']=dataset['Dep_Time'].map({'Class 1':1 , 'Class 2':2 , 'Class 3':3 , 'Class 4': 4})
dataset['Arrival_Time']=dataset['Arrival_Time'].map({'Class 1':1 , 'Class 2':2 , 'Class 3':3 , 'Class 4': 4})
```

```python
dataset.head()
```

```python
X=dataset
Y=dataset['Price']
X=X.drop('Price',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from xgboost import XGBRegressor
reg=XGBRegressor()
reg.fit(X_train,Y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
for_reg=RandomForestRegressor()
for_reg.fit(X_train,Y_train)
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada_reg=AdaBoostRegressor()
ada_reg.fit(X_train,Y_train)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb_reg=GradientBoostingRegressor()
gb_reg.fit(X_train,Y_train)
```

```python
model1_output=reg.predict(X_test)
model2_output=for_reg.predict(X_test)
model3_output=ada_reg.predict(X_test)
model4_output=gb_reg.predict(X_test)
```

```python
import math
from sklearn.metrics import mean_squared_error
a1=mean_squared_error(Y_test,model1_output)
a2=mean_squared_error(Y_test,model2_output)
a3=mean_squared_error(Y_test,model3_output)
a4=mean_squared_error(Y_test,model4_output)
print('XGB : {}'.format(math.sqrt(a1)))
print('Random Forest : {}'.format(math.sqrt(a2)))
print('ADA : {}'.format(math.sqrt(a3)))
print('GB : {}'.format(math.sqrt(a4)))
```
